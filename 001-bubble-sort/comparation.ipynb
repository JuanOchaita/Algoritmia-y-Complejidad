# Bubble sort: baseline vs Optimized

### Baseline bubble sort:

```python

def bubble_sort(arr):

    n = len(arr)

    for _ in range(n):
        for j in range(0, n - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr
```

### Optimized bubble sort

```python

def opt_bubble_sort(arr):

    n = len(arr)

    for i in range(n):

        is_sorted = True
        
        for j in range(0, n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                is_sorted = False

        if is_sorted:
            return arr

    return arr
```

#### Time complexity


```python
import time

def time_function(func, *args, **kwargs):
    start_time = time.perf_counter()
    result = func(*args, **kwargs)
    end_time = time.perf_counter()
    duration = end_time - start_time
    print(f'Execution time: {duration:.10f} secs')
    return duration

```

#### List of 5k, 10k, 25k and 50k numbers (BIGO)

```python
numbers_05k = [i for i in range(5000, 0, -1)]
numbers_10k = [i for i in range(10000, 0, -1)]
numbers_25k = [i for i in range(25000, 0, -1)]
numbers_50k = [i for i in range(50000, 0, -1)]

bubble_sort_times = [0]
opt_bubble_sort_times = [0]

print(numbers_05k)

print(bubble_sort(numbers_05k.copy()))
```

#### Baseline bubble sort:

```python
bubble_sort_times.append(time_function(bubble_sort, numbers_05k.copy()))
bubble_sort_times.append(time_function(bubble_sort, numbers_10k.copy()))
bubble_sort_times.append(time_function(bubble_sort, numbers_25k.copy()))
bubble_sort_times.append(time_function(bubble_sort, numbers_50k.copy()))

print(bubble_sort_times)
```

#### Optimized bubble sort

```python
opt_bubble_sort_times.append(time_function(opt_bubble_sort, numbers_05k.copy()))
opt_bubble_sort_times.append(time_function(opt_bubble_sort, numbers_10k.copy()))
opt_bubble_sort_times.append(time_function(opt_bubble_sort, numbers_25k.copy()))
opt_bubble_sort_times.append(time_function(opt_bubble_sort, numbers_50k.copy()))

print(opt_bubble_sort_times)
```

#### Graph

```python
import matplotlib.pyplot as plt

bubble_sort_times_float =  [float(item) for item in bubble_sort_times]
opt_bubble_sort_times_float =  [float(item) for item in opt_bubble_sort_times]

print(f"{bubble_sort_times}")
print(f"{opt_bubble_sort_times}")

x = ["0","5K", "10K", "25K", "50K"]

plt.figure(figsize=(8, 5))
plt.plot(x, bubble_sort_times_float, marker='o', color='blue', label='Baseline Bubble Sort Times')
plt.plot(x, opt_bubble_sort_times_float, marker='o', color='red', label='Optimized Bubble Sort Times')
plt.title('Tiempos bubble sort: baseline vs Optimized')
plt.xlabel('Cantidad de numeros')
plt.ylabel('Tiempo (segundos)')
plt.legend()
plt.grid(True)
plt.show()
```
